```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_df = pd.merge(student_data, school_data, how="left", on="school_name")
school_data_df.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# calculate the total number of schools
schools = school_data_df["school_name"].unique()
number_schools = len(schools)
number_schools
```

```python
# calculate the total number of students
students = school_data_df["Student ID"].unique()
number_students = len(students)
number_students
```

```python
# calculate the total budget for the district
grouped_schools_df = school_data_df.groupby(["school_name"])
total_budget = int(grouped_schools_df["budget"].unique().sum())
total_budget 


```

```python
# calculate the average math score
average_math = school_data_df["math_score"].mean()
average_math
```

```python
# calculate the average reading score
average_reading = school_data_df["reading_score"].mean()
average_reading
```

```python
# calculate the % of students with a passing math score (70 or greater)
# number passing/number students
passing_students_df = school_data_df.loc[(school_data_df["math_score"] >= 70)]
passing_students_len = len(passing_students_df)
percent_passing_math = (passing_students_len/number_students)*100
percent_passing_math
```

```python
# calculate the % of students with a passing reading score (70 or greater)
# number passing/number students
passing_reading_df = school_data_df.loc[(school_data_df["reading_score"] >= 70)]
passing_reading_len = len(passing_reading_df)
percent_passing_reading = (passing_reading_len/number_students)*100
percent_passing_reading
```

```python
#calculate the % of students who passed math AND reading
pass_both_df = school_data_df.loc[(school_data_df["reading_score"] >= 70) & (school_data_df["math_score"] >= 70)]
number_students_pass_both = len(pass_both_df)
percent_students_pass_both = (number_students_pass_both / number_students)*100
percent_students_pass_both
```

```python
#create the dataframe from the above variables
district_summary_df = pd.DataFrame({"Total Number of Schools": [number_schools], "Total Number of Students": [number_students], 
                           "Total Budget": [total_budget], "Average Math Score": [average_math], 
                           "Average Reading Score": [average_reading], "% Math Passed": [percent_passing_math], 
                           "% Reading Passed": [percent_passing_reading], "% Passed Both": [percent_students_pass_both]})

#format the dataframe
district_summary_df["Total Number of Students"] = district_summary_df["Total Number of Students"].astype(float).map("{:,.0f}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].astype(float).map("${:,.0f}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].astype(float).map("{:,.2f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].astype(float).map("{:,.2f}".format)
district_summary_df["% Math Passed"] = district_summary_df["% Math Passed"].astype(float).map("% {:,.2f}".format)
district_summary_df["% Reading Passed"] = district_summary_df["% Reading Passed"].astype(float).map("% {:,.2f}".format)
district_summary_df["% Passed Both"] = district_summary_df["% Passed Both"].astype(float).map("% {:,.2f}".format)

district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# create boolean columns for passing math, passing reading, and passing both within the original dataframe
school_data_df["Passing Math"] = school_data_df["math_score"] >= 70
school_data_df["Passing Reading"] = school_data_df["reading_score"] >= 70
school_data_df["Passing Both"] =  school_data_df.apply(lambda x: x['math_score'] >= 70 and x['reading_score'] >= 70, axis=1)

school_data_df
```

```python
# group the dataframe by school name and take the average of numerical columns
school_grouped_df = school_data_df.groupby(["school_name"]).mean()

# add a column for per student budget by taking the quotient of total school budget and total students
school_grouped_df["Per Student Budget"] = school_grouped_df["budget"]/school_grouped_df["size"]
school_grouped_df

#merge school_grouped_df with school_data to get information about type
merged_df = pd.merge(school_grouped_df, school_data, how="left", on="school_name")

# delete extraneous columns
del merged_df["Student ID"]
del merged_df["School ID_x"]
del merged_df["School ID_y"]
del merged_df["size_x"]
del merged_df["budget_x"]


#create DataFrame from merged_df
school_summary_df = pd.DataFrame({"School Name": merged_df["school_name"], "School Type": merged_df["type"], 
                                  "Total Students": merged_df["size_y"], "Total School Budget": merged_df["budget_y"], 
                                  "Per Student Budget": merged_df["Per Student Budget"], 
                                  "Average Math Score": merged_df["math_score"], 
                                  "Average Reading Score": merged_df["reading_score"],
                                  "% Passing Math": (merged_df["Passing Math"])*100, 
                                  "% Passing Reading": (merged_df["Passing Reading"])*100, 
                                  "% Overall Passing": (merged_df["Passing Both"])*100})
#set index to school name
school_summary_df.set_index("School Name", inplace=True)

# see bottom of notebook for formatted output
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# create pandas series' for each grade and groupby schoool
ninth_grade = school_data_df[school_data_df["grade"] == "9th"].groupby("school_name").mean()["math_score"]
tenth_grade = school_data_df[school_data_df["grade"] == "10th"].groupby("school_name").mean()["math_score"]
eleventh_grade = school_data_df[school_data_df["grade"] == "11th"].groupby("school_name").mean()["math_score"]
twelfth_grade = school_data_df[school_data_df["grade"] == "12th"].groupby("school_name").mean()["math_score"]

# create a dataframe from the pandas series created above
math_scores_grade_df = pd.DataFrame({"9th": ninth_grade, "10th": tenth_grade, 
                                     "11th": eleventh_grade, "12th": twelfth_grade})
#format the dataframe
math_scores_grade_df["9th"] = math_scores_grade_df["9th"].astype(float).map("{:,.2f}".format)
math_scores_grade_df["10th"] = math_scores_grade_df["10th"].astype(float).map("{:,.2f}".format)
math_scores_grade_df["11th"] = math_scores_grade_df["11th"].astype(float).map("{:,.2f}".format)
math_scores_grade_df["12th"] = math_scores_grade_df["12th"].astype(float).map("{:,.2f}".format)



math_scores_grade_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# create bins
bins = [0, 584.99, 629.99, 644.99, 679.99]

# create bin labels
labels = ["<$585", "$585-630", "$630-645", "$645-680"]


#create spending ranges column and bin data from the school summary df
school_summary_df["Spending Ranges (Per Student)"] = pd.cut(school_summary_df["Per Student Budget"], 
                                                            bins, labels=labels, include_lowest=True)

#groupby spending ranges 
grouped_spending_ranges = school_summary_df.groupby("Spending Ranges (Per Student)")

#assign groupby object to spending_range_df and take averages of numerical columns
spending_range_df = grouped_spending_ranges.mean()

#delete extraneous columns from spending_range_df
del spending_range_df["Total Students"]
del spending_range_df["Total School Budget"]
del spending_range_df["Per Student Budget"]


#format the dataframe
spending_range_df["Average Math Score"] = spending_range_df["Average Math Score"].astype(float).map("{:,.2f}".format)
spending_range_df["Average Reading Score"] = spending_range_df["Average Reading Score"].astype(float).map("{:,.2f}".format)
spending_range_df["% Passing Math"] = spending_range_df["% Passing Math"].astype(float).map("%{:,.2f}".format)
spending_range_df["% Passing Reading"] = spending_range_df["% Passing Reading"].astype(float).map("%{:,.2f}".format)
spending_range_df["% Overall Passing"] = spending_range_df["% Overall Passing"].astype(float).map("%{:,.2f}".format)


spending_range_df


```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# create bins
size_bins = [0, 999, 1999, 5000]

# create bin labels
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

#create school size column and bin data from the school summary df
school_summary_df["School Size"] = pd.cut(school_summary_df["Total Students"], 
                                                            size_bins, labels=size_labels, include_lowest=True)

#groupby school sizes 
grouped_school_sizes = school_summary_df.groupby("School Size")

#assign groupby object to spending_range_df
school_sizes_df = grouped_school_sizes.mean()

# #delete extraneous columns from spending_range_df
del school_sizes_df["Total Students"]
del school_sizes_df["Total School Budget"]
del school_sizes_df["Per Student Budget"]

#format the dataframe
school_sizes_df["Average Math Score"] = school_sizes_df["Average Math Score"].astype(float).map("{:,.2f}".format)
school_sizes_df["Average Reading Score"] = school_sizes_df["Average Reading Score"].astype(float).map("{:,.2f}".format)
school_sizes_df["% Passing Math"] = school_sizes_df["% Passing Math"].astype(float).map("%{:,.2f}".format)
school_sizes_df["% Passing Reading"] = school_sizes_df["% Passing Reading"].astype(float).map("%{:,.2f}".format)
school_sizes_df["% Overall Passing"] = school_sizes_df["% Overall Passing"].astype(float).map("%{:,.2f}".format)

school_sizes_df
```
